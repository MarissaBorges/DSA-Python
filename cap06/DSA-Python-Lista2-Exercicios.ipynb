#### Exercício 1

Escreva uma função que receba os valores de três lados de um triângulo e o classifique como "Equilátero" (todos os lados iguais), "Isósceles" (dois lados iguais) ou "Escaleno" (todos os lados diferentes).

```python
# Solução
exibir_info('BEM VINDO AO VERIFICADOR DE TRIÂNGULOS')

print("\nInforme os valores dos 3 lados do triângulo")
valor1 = int(input("Valor 1: "))
valor2 = int(input("Valor 2: "))
valor3 = int(input("Valor 3: "))

condicao_existencia1 = (valor1 + valor2) > valor3
condicao_existencia2 = (valor1 + valor3) > valor2
condicao_existencia3 = (valor2 + valor3) > valor1

if condicao_existencia1 and condicao_existencia2 and condicao_existencia3:
    print(f"Esse triângulo com os valores: {valor1, valor2, valor3} é ", end="")
    if valor1 == valor2 and valor2 == valor3:
        print("Equilátero")
    elif valor1 != valor2 and valor2 != valor3:
        print("Escaleno")
    else:
        print("Isóceles")
else:
    print("Esses valores não formam um triângulo!!")
```

#### Exercício 3 

Você recebeu um dicionário com os nomes dos alunos e suas respectivas notas. Escreva uma função que calcula a média da turma e retorna uma lista com os nomes dos alunos que tiveram nota acima da média.

```python
# Solução
turma = {
    "Pedro": 10,
    "Maria": 5.5,
    "Lucas": 4,
    "Eduarda": 9,
    "Fabio": 7,
    "Eduardo": 0,
    "Diego": 2
}

def alunos_acima_media(turma):
    """Calcula a média da turma e retorna um dicionário com os alunos acima da média"""
    if not turma:
        print("Dicionário de turma vazio")
        return {}

    media = sum(turma.values()) / len(turma)
    print(f"A média da turma é {media:.2f}")

    alunos_acima_media = [x for x in turma.items() if x[1] > media]

    return alunos_acima_media

exibir_info("ALUNOS ACIMA DA MÉDIA")
for aluno in alunos_acima_media(turma):
    print(f"Aluno: {aluno[0]}, Nota: {aluno[1]}")
```

#### Exercício 5

Crie uma função chamada dsa_calcula_imc que aceite dois argumentos: peso (em kg) e altura (em metros). A função deve calcular o Índice de Massa Corporal (IMC) usando a fórmula IMC=peso/altura^2 e retornar o valor do IMC.

```python
# Solução
def dsa_calcula_imc(peso, altura):
    if altura <= 0:
        print("Valor de altura é inválido")
        return 0
    else:
        return peso / altura ** 2

imc = dsa_calcula_imc(62, 1.65)
print(f"IMC: {imc:.2f}")
```

#### Exercício 7

Crie uma função que receba uma lista de números inteiros e retorne um dicionário contendo a contagem de quantos números são pares e quantos são ímpares.

```python
# Solução
numeros = [445,54,45,34324,452,56,869,987,642,75,87,98,98,0,97,87,543,3,54,65,8,9,87,7,4,5,6,7,68,78,7,765,6,435,32,]

def contar_numeros(numeros: list[int]):
    numeros_pares = [x for x in numeros if x % 2 == 0]
    numeros_impares = [x for x in numeros if x % 2 != 0]
    dict_numeros = {"Pares": {"Números": sorted(numeros_pares), "Contagem": len(numeros_pares)}, "Impares": {"Números": sorted(numeros_impares), "Contagem": len(numeros_impares)}}
    return dict_numeros

print(contar_numeros(numeros))
```

#### Exercício 8

Crie uma função que receba uma lista de strings (potenciais e-mails) e um parâmetro opcional dominio_desejado (com valor padrão "gmail.com"). A função deve retornar uma nova lista, usando list comprehension, contendo apenas os e-mails que terminam com o domínio desejado.

```python
# Solução
emails = [
    "carlos@gmail.com",
    "ana.v@hotmail.com",
    "admin@empresa.com.br",
    "suporte@gmail.com",
    "marta@uol.com.br",
    "ti@empresa.com.br",
    "contato@gmail.com",
    "vendas@outlook.com",
    "rh@empresa.com.br",
    "joao@uol.com.br",
    "felipe@gmail.com",
    "diretoria@empresa.com.br",
    "luiza@hotmail.com",
    "dev@gmail.com",
    "webmaster@uol.com.br",
]

def verificar_dominio(emails, dominio = "gmail.com"):
    emails_split = [e.split("@") for e in emails]
    dominios = [d for d in emails_split if d[1] == dominio]
    emails_validos = [{e[0]+"@"+e[1]} for e in dominios]
    return emails_validos

print(f"Emails Válidos: {verificar_dominio(emails)}")

######################## VERSAO CLEAN CODE
#def verificar_dominio(emails, dominio="gmail.com"):
#    # Filtra e já formata o resultado final de uma vez
#    emails_validos = [e for e in emails if e.endswith(f"@{dominio}")]
#    return emails_validos
```

#### Exercício 10

Crie um jogo onde o computador escolhe um número secreto entre 1 e 20. O jogador tem 5 tentativas para adivinhar. A cada tentativa, o programa informa se o palpite foi muito alto ou muito baixo. Se o jogador acertar, o loop deve ser interrompido com uma mensagem de vitória.

```python
# Solução
import random

exibir_info("JOGO DA ADVINHAÇÃO")

print("\nRegras do Jogo: \n - Você tem 5 tentativas para acertar o número que o Python escolheu\n - Em cada chute você terá um retorno dizendo se foi alto ou baixo\n\n Boa Sorte!!\n")
numero_aleatorio = random.randint(1, 20)

while True:
    valor = int(input("Seu chute: "))
    if valor == numero_aleatorio:
        print("Parabéns!!! Você acertou!!!")
        break
    elif valor > numero_aleatorio:
        print("Seu chute foi alto")
    else:
        print("Seu chute foi baixo")
```
